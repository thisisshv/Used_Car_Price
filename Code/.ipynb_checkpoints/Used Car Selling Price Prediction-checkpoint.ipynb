# Used Car Price Prediction

### Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

### Loading data

```python
df = pd.read_csv("D:/Code/Projects/Used_Car_Price/Dataset/used_car_dataset.csv")
df
```

### EDA

```python
df.isna().sum()
```

```python
# Dropping rows with null value

df = df.dropna(axis=0)
df
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
# Removing ₹ sign
df['AskPrice'] = df['AskPrice'].str.replace('₹', '', regex=False)
# Removing commas
df['AskPrice'] = df['AskPrice'].str.replace(',', '', regex=False)
# Convert to int
df['AskPrice'] = df['AskPrice'].astype(int)
```

```python
# Removing km
df['kmDriven'] = df['kmDriven'].str.replace('km', '', regex=False)
# Removing commas
df['kmDriven'] = df['kmDriven'].str.replace(',', '', regex=False)
# Comvert to int
df['kmDriven'] = df['kmDriven'].astype(float)
```

```python
df.info()
```

```python
# Changing Age from years to month for better accuracy
df['Age'] = df['Age']*12
df
```

```python
# Dropping unnecessary columns
df = df.drop(['Year', 'PostedDate', 'AdditionInfo'], axis=1)
df
```

```python
df = df.rename(columns={'model':'Model'})
df = df.rename(columns={'kmDriven':'KmDriven'})
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
# Dropping duplicate values

df = df.drop_duplicates()
df
```

```python
df.info()
```

### Encoding

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df['Brand'] = encoder.fit_transform(df['Brand'])
df['Model'] = encoder.fit_transform(df['Model'])
df['Transmission'] = encoder.fit_transform(df['Transmission'])
df['Owner'] = encoder.fit_transform(df['Owner'])
df['FuelType'] = encoder.fit_transform(df['FuelType'])
```

```python
df
```

### Splitting into Train, Test

```python
from sklearn.model_selection import train_test_split

x = df.drop('AskPrice', axis=1)
y = df['AskPrice']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

print("Training set shape:", x_train.shape)
print("Testing set shape:", x_test.shape)
```

```python
x_train
```

### Model Training & Evaluation

#### Model Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score

def metric_calc(x, y):
    y_pred = model.predict(x)
    y_true = y
    
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = root_mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    # adjusted r2
    # number of values in the data set
    n = y_true.size
    # number of independent features
    k = 7
    adjusted_r2 = 1 - ((1-r2) * ((n - 1) / (n - (k + 1))))
    adjusted_r2
    
    print(f"Metrics:\n mae = {mae}\n mse = {mse}\n rmse = {rmse}\n r2 = {r2}\n adjusted_r2 = {adjusted_r2}")
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# training accuracy
y_pred = model.predict(x_train)
y_true = y_train

print(f"Training")
metric_calc(y_pred, y_true)

# testing accuracy
y_pred = model.predict(x_test)
y_true = y_test

print(f"Testing")
metric_calc(y_pred, y_true)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(x_train, y_train)
```

```python
# training accuracy
y_pred = model.predict(x_train)
y_true = y_train

print(f"Training")
metric_calc(y_pred, y_true)

# testing accuracy
y_pred = model.predict(x_test)
y_true = y_test

print(f"Testing")
metric_calc(y_pred, y_true)
```
